## 1 A simple plot

### 1.1 Letâ€™s look at some code.

```python
from matplotlib import pyplot as plt
```

```python
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

plt.plot(x, y_1, color='blue', linestyle='dotted', label='Y values')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=0.30)
plt.legend(loc='lower right')
plt.show()
```

```python
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

plt.plot(x, y_1, color='magenta', linestyle='dashed', linewidth=2)
```

### 1.2 Adding another plot

```python
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

plt.plot(x, y_1, c='orange', ls='dashed', lw=2, label = 'Y1 values')
plt.plot(x,y_2, c='indigo', ls='dashed', lw=2, label = 'Y2 values')

plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=0.30)
plt.legend(loc='upper left')
plt.show()
```

### 1.3 Yet another plot but with error bars

```python
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = 0.1, 0.75

plt.plot(x, y_1, c='#f6bd60', ls='dashed', lw=2, label = 'Y1 values')
plt.plot(x,y_2, c='#84a59d', ls='dashed', lw=2, label = 'Y2 values')
plt.errorbar(x, y_3, xerr = x_error, yerr = y_error, ls='dashed', lw=2,
             color = "#f5cac3", label = "Y3 with errors")

plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.legend(loc='upper left')
plt.grid(alpha = 0.5, color = '#fec89a', linestyle = 'dotted')
plt.show()
```

### 1.5 Better with NumPy

```python
import numpy as np
```

```python
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

np_x = np.array(x)
np_y_1 = np.array(y_1)
np_y_2 = np.array(y_2)

plt.plot(np_x, np_y_1, color='#c1121f', linestyle='dashed', label='Y values')
plt.plot(np_x, np_y_2, color='#669bbc', linestyle = 'dashed', label='Y2 values')
plt.show()
```

### 1.6 Adding mathematical functions

```python
x = np.array([0,1,2,3,4,5,6])
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, color='#9c89b8', linestyle='dashed', label='x^2')
plt.plot(x, sin_x, color='#f0a6ca', label='sin(x)')
plt.legend(loc='upper left')
plt.show()
```

### 1.7 We need another axis!

```python
x = np.linspace(0, 6, 100)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, color='#9c89b8', linestyle='dashed', label='x^2')
plt.legend(loc='lower left')

plt.twinx()
plt.plot(x, sin_x, color='#f0a6ca', label= 'sin(x)')
plt.legend(loc='lower right')
plt.show()
```

### 1.8 Saving to disc

```python
x = np.linspace(0, 6, 100)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, color='#669bbc', linestyle='dashed', label='x^2')
plt.legend(loc='lower left')

plt.twinx()
plt.plot(x, sin_x, color='#c1121f', label='sin(x)')
plt.legend(loc='lower right')

plt.savefig('simple-plot.png', dpi=150)
```

## 2 A real example: Global Warming

### 2.1 Plotting data from files

```python
data = np.loadtxt('Complete_TAVG_daily.txt', skiprows=24)
date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.ylim([-8,8])
```

```python
data = np.loadtxt('Complete_TAVG_daily.txt', skiprows=24)
date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.ylim([-8, 8])
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
```

```python
plt.style.use('fivethirtyeight')

data = np.loadtxt('Complete_TAVG_daily.txt', skiprows=24)
date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.ylim([-8, 8])
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')

plt.xlabel('Year')
plt.ylabel('Temperature Anomaly')
plt.title('Temperature anomaly \n(Relative to average from Jan 1951 - Dec 1980)')
plt.grid(alpha=0.30)
```
